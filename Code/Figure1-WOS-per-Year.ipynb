## Figure 1 CRANE - Trend Lines WOS

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

### Import CSV data

```python
data_import = pd.read_csv('~/Documents/CNS/CRANE/Figures/Figure1-Totals/wos-per-year-1900-2017-04-01-19.csv', 
                          encoding='latin1', sep=",")
```

```python
list(data_import)
```

### Ploting Trend Line

```python
plt.style.use('seaborn-paper')
```

```python
plt.rcParams['svg.fonttype'] = 'none' # for svg formatting
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
plt.rcParams['font.sans-serif'] = "Arial"

plt.rcParams['font.family'] = "sans-serif"

np.seterr(divide='ignore', invalid='ignore') # to remove warning messages
fig = plt.figure(figsize=(12, 6),facecolor='white') # gigure size and white background
x = data_import['Year'].values # year values - change name if the column has a different name
ax = plt.gca()
#ax1.xaxis.set_major_locator(MaxNLocator(integer=True)) # improve axis ticks
y = data_import['Count'].values
plt.yscale('log')  # uncomment if log values are used
ax.plot(x, y, linewidth=2.0, color='red') # select color
ax.set_xticks([1900,  1920,1940,1960,1980, 2000, 2017])
ax.set_yticks([10000,100000, 1000000])
ax.yaxis.set_ticks_position('none') 
ax.set_xlabel('Year',  fontsize=13) # fontname="Arial"
ax.set_ylabel('#Papers',  fontsize=13) # for log values use ax1.set_ylabel('Log(#Papers)')

plt.savefig('wos-per-year-total-1990-2017.png')
plt.savefig('wos-per-year-total-1990-2017.svg')
plt.show()
```
