## Figure 1C CRANE - WOS/NSF Trendlines

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

### Import CSV data

```python
wospub = pd.read_csv('wos-all-merged-year-counts.csv', encoding='latin1', sep=",")
nsfpub = pd.read_csv('nsf-all-merged-year-counts-1998.csv', encoding='latin1', sep=",")
woskey = pd.read_csv('wos-all-year-keywords-counts.csv', encoding='latin1', sep=",")
nsfkey = pd.read_csv('nsf-all-year-keywords-counts.csv', encoding='latin1', sep=",")
```

### Ploting Trend Line

```python
plt.style.use('seaborn-paper')
```

```python
plt.rcParams['axes.xmargin'] = 0 # remove whote margins
plt.rcParams['svg.fonttype'] = 'none' # for svg formatting
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
plt.rcParams['font.sans-serif'] = "Arial"

plt.rcParams['font.family'] = "sans-serif"

np.seterr(divide='ignore', invalid='ignore') # to remove warning messages
fig = plt.figure(figsize=(12, 6),facecolor='white') # gigure size and white background

x1 = wospub.Year.values
ax1 = plt.gca()
#ax1 = fig.add_subplot(121)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1.plot(x1, wospub.Count.values, linewidth=2.0, color='black') # 
ax1.plot(x1, nsfpub.Count.values, linewidth=2.0, color='green') # #4daf4a green

ax1.plot( x1,nsfkey.Count.values, linewidth=2.0, color='darkblue') ##0707d2 darkblue
# Make the y-axis label, ticks and tick labels match the line color.
ax1.plot(x1, woskey.Count.values, linewidth=2.0, color='darkred') # red ##8d1b18 dark read

ax1.set_ylabel('Count',fontsize=13)

ax1.legend(['WOS Publications (AI+Robotics+IoT)','NSF Awards (AI+Robotics+IoT)', 'NSF Keywords (AI+Robotics+IoT)', 'WOS Keywords (AI+Robotics+IoT)'], 
           loc='upper center',fontsize=13)

ax1.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.show()
```
